```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as Stats
```

```python
cars=pd.read_csv('Cars.csv')
```

```python
cars
```

```python
cars.mean()
```

```python
cars.median()
```

```python
cars.mode()
```

```python
cars.rename(columns={'HP':'hp'},inplace= True)
```

```python
cars
```

```python
cars.describe()
```

```python
plt.hist(cars["hp"],bins=20,edgecolor='black')
```

```python
plt.boxplot(x='hp',data=cars)
plt.xlabel('hp')
plt.ylabel('wt')
plt.savefig("pointsinference.png")
plt.show()
```

```python

plt.figure(figsize=(16,9))
plt.barh(cars['hp'],cars['MPG'])
plt.yticks(fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(cars.index)
plt.xlabel('hp')
plt.ylabel('wt')
plt.show()
```

```python

sns.distplot(cars.index,label='Cars.csv')
plt.xlabel('hp')
plt.ylabel('wt')
plt.legend()
plt.show()
```

```python
sns.scatterplot(cars.index,label='Cars.csv')
plt.xlabel('hp')
plt.ylabel('wt')
plt.legend()
plt.show()          
```

```python
plt.hist(cars["MPG"], bins = 20, edgecolor=  'black')
plt.show()
```

```python
plt.boxplot(x= 'MPG', data =cars)
plt.show()
```

```python
sns.distplot(cars['MPG'],kde=True, bins =10)
plt.show()
```

```python
import statsmodels.api as sm
sm.qqplot(cars['MPG'])
plt.xlabel('MPG', color ='blue')
plt.savefig('MPG of cars.png')
plt.show()
```

```python
import scipy.stats as stats
stats.probplot(cars['MPG'], dist="norm", plot=plt)
plt.xlabel('MPG', color ='blue')
plt.savefig('MPG of cars.png')
plt.show()
```
